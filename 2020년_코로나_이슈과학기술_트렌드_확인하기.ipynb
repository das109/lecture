```python
def remove_html(sentence):
	sentence = re.sub('(<([^>]+)>)', '', sentence) 
	sentence = sentence.replace("\n", " ") 
	sentence = sentence.replace("@media all and (min-width: 1px) and (max-width: 450px) { .dable_placeholder{ width: 100% !important; padding: 10px 0 !important; }}",  "")
	sentence = sentence.replace("#AD165415868344.ad-template .col { text-align:center; }",  "")
	sentence = sentence.replace("#AD165415868344.ad-template .col .ad-view { position:relative; }",  "")
	sentence = sentence.replace("#AD165415868344.ad-template { margin:auto; position:relative; display:block; clear:both; z-index:1; }",  "")

	sentence = sentence.strip()
 
	return sentence
```

```python
import re

# 텍스트 정제 함수 : 한글 이외의 문자는 전부 제거합니다.
def text_cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글의 정규표현식을 나타냅니다.
    result = hangul.sub('', text)
    return result
```

```python
#  피처에 이를 적용합니다.
df['기사제목'] = df['기사제목'].apply(lambda x: text_cleaning(x))
df['기사부제목'] = df['기사부제목'].apply(lambda x: text_cleaning(x))
df['기사내용'] = df['기사내용'].apply(lambda x: text_cleaning(x))
df.head()
```

```python
from konlpy.tag import Okt
from collections import Counter

# 한국어 약식 불용어사전 출처 - (https://www.ranks.nl/stopwords/korean)
korean_stopwords_path = "/content/drive/MyDrive/data/korean_stopwords.txt"
with open(korean_stopwords_path, encoding='utf8') as f:
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]

def get_nouns(x):
    nouns_tagger = Okt()
    nouns = nouns_tagger.nouns(x)
    
    # 한글자 키워드를 제거
    nouns = [noun for noun in nouns if len(noun) > 1]
    
    # 불용어를 제거
    nouns = [noun for noun in nouns if noun not in stopwords]
    
    return nouns
```

```python
from konlpy.tag import Okt
from collections import Counter
from apyori import apriori
```

```python
# 트랜잭션 데이터를 추출합니다.
df_tmp = df_2020_li[4]
transactions = df_tmp['명사들'].tolist()
transactions = [transaction for transaction in transactions if transaction] # 공백 문자열을 방지합니다.
print(transactions)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# 연관 분석을 수행합니다.
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
transactions_df = pd.DataFrame(te_ary, columns=te.columns_)

transactions_df.head()
```

```python
frequent_itemset = apriori(transactions_df,
                           min_support=0.05, 
                           max_len=2, 
                           use_colnames=True,
                          )
frequent_itemset = pd.DataFrame(frequent_itemset)
  
  
association_rules_df = association_rules(frequent_itemset, 
                                         metric='confidence', 
                                         min_threshold=0.005,
                                        )
df_sorted = association_rules_df.sort_values(by='support',ascending=False)
df_transactions = association_rules_df.sort_values(by='support',ascending=False)

df_sorted.head()
```

```python
columns = ['source', 'target', 'support']
network_df = pd.DataFrame(columns=columns)

network_df = df_transactions[['antecedents','consequents', 'support']]
network_df.columns = ['source', 'target', 'support']
print(network_df)


network_df['source'] = network_df['source'] .astype(str)
network_df['source'] = network_df['source'].str.slice(start=12, stop=14)
network_df['target'] = network_df['target'] .astype(str)
network_df['target'] = network_df['target'].str.slice(start=12, stop=14)

print(network_df)
```

```python
columns = ['source', 'target', 'support']
network_df = pd.DataFrame(columns=columns)

network_df = df_transactions[['antecedents','consequents', 'support']]
network_df.columns = ['source', 'target', 'support']


network_df['source'] = network_df['source'] .astype(str)
network_df['source'] = network_df['source'].str.slice(start=12, stop=14)
network_df['target'] = network_df['target'] .astype(str)
network_df['target'] = network_df['target'].str.slice(start=12, stop=14)


transactions_final = []
for i in range(len(transactions)):
    transactions_final = transactions_final + transactions[i]


count_items = Counter(transactions_final)
print(count_items)

node_df = pd.DataFrame(count_items.items(), columns=['node', 'nodesize'])

node_df = node_df.sort_values(by=['nodesize'], ascending=False)
node_df = node_df[0:30]
```

```python
network_df1 = pd.DataFrame(columns=network_df.columns)
node_list = np.array(node_df['node'].tolist())
print(node_list)
for index, row in network_df.iterrows():
    if (row["source"] in node_list) & (row["target"] in node_list): 
        network_df1.loc[index] = row

network_df1 = network_df1.sort_values(by=['support'], ascending=False)
network_df = network_df1[0:500]
```

```python
import networkx as nx

plt.figure(figsize=(10,10))

# networkx 그래프 객체를 생성합니다.
G = nx.Graph()

# node_df의 키워드 빈도수를 데이터로 하여, 네트워크 그래프의 ‘노드’ 역할을 하는 원을 생성합니다.
for index, row in node_df.iterrows():
      G.add_node(row['node'], nodesize=row['nodesize'])
    
# network_df의 연관 분석 데이터를 기반으로, 네트워크 그래프의 ‘관계’ 역할을 하는 선을 생성합니다.
for index, row in network_df.iterrows():
      G.add_weighted_edges_from([(row['source'], row['target'], row['support'])])
G.remove_nodes_from(list(nx.isolates(G)))
# 그래프 디자인과 관련된 파라미터를 설정합니다.
pos = nx.spring_layout(G, k=0.6, iterations=50)
sizes = [G.nodes[node]['nodesize']*25 for node in G]
nx.draw(G, pos=pos, node_size=sizes)

# Windows 사용자는 AppleGothic 대신,'Malgun Gothic'. 그 외 OS는 OS에서 한글을 지원하는 기본 폰트를 입력합니다.
nx.draw_networkx_labels(G, pos=pos, font_family=font_name, font_weight='bold', font_size=15)

# 그래프를 출력합니다.
ax = plt.gca()
plt.show()
```
