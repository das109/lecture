# 1. 네트워크 이론을 통한 산업 내 주요 기업, 혹은 라이벌 기업 분석

```python
search_input = input("인도에서 판매하고자 하는 제품을 입력해주세요.: ")
print("")
print("")
search_result = search(search_input)

search_result = search_result[['CONSIGNEE', 'SHIPPER', 'QUANTITY']]
search_result.sort_values('QUANTITY', ascending=False).head(30)
```

노드 종류  
1. eigenvector centrality: 영향력이 강한 기업과 연결된 기업을 확인 가능. 그러나 허용오차 수렴에 실패하여 오류 발생.  
2. betweenness centrality: 네트워크의 기업들이 해당 기업을 지나가는지 고려한 네트워크. 만약 기업들이 해당 기업을 지나간다면 해당 기업의 betweenness centrality가 높음.  
3.degree centrality: 각 기업별로 직접 연결된 기업만을 고려. 즉, 해당 기업이 가진 영향력의 크기만을 고려.


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

import networkx as nx


G = nx.DiGraph()
for tup in search_result[['CONSIGNEE','SHIPPER']].itertuples():
    G.add_edge(tup.CONSIGNEE, tup.SHIPPER)
    
print(len(search_result))
print(len(G))     
pos = nx.spring_layout(G)


# 참고: 1. eigenvector centrality의 경우 허용오차 수렴에 실패 (이 경우 오류가 발생함)

# 2. betweenness_centrality 계산
# 네트워크의 모든 노드 쌍 간의 shortest path가 해당 노드를 지나는지를 고려한 centrality
# 전체 네트워크의 흐름을 제어하며, 해당 노드가 없어질 경우, 전체 네트워크의 흐름 자체에 영향을 받음

betCent = nx.betweenness_centrality(G, normalized=True, endpoints=True)
#node_color = [-0.1 * G.degree(v) for v in G]

d = dict(G.degree)
low, *_, high = sorted(d.values())
norm = mpl.colors.Normalize(vmin=low*1.5, vmax=high*1.5, clip=True)
mapper = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.Blues)

node_size =  [v * 700000 for v in betCent.values()]
plt.figure(figsize=(20,20))

G_rev = nx.DiGraph.reverse(G)

nx.draw_networkx(G_rev, pos=pos, with_labels=True,
                 node_color=[mapper.to_rgba(i) for i in d.values()],
                 node_size=node_size, font_size=8)


plt.axis('off')
```

```python
# degree centrality 계산
# node별로 직접 연결된 edge의 weight를 고려, weight는 betweenness_centrality 고려   
# 해당 node가 직접 가진 영향력의 크기는 얼마인지 측정

N_nodes = len(G)

# degree_centrality 계산
degCent = nx.degree_centrality(G)
degCent_sorted=dict(sorted(degCent.items(), key=lambda item: item[1], reverse=True))

# betweenness_centrality 계산
betCent = nx.betweenness_centrality(G, normalized=True, endpoints=True)
betCent_sorted=dict(sorted(betCent.items(), key=lambda item: item[1], reverse=True))

color_list=N_nodes*['lightsteelblue']
N_top=10
colors_top_10=['tab:blue','lightsteelblue']
keys_deg_top=list(degCent_sorted)[0:N_top]
keys_bet_top=list(betCent_sorted)[0:N_top]


inter_list=list(set(keys_deg_top) & set(keys_bet_top))

for i in range(len(inter_list)):
  color_list[i]=colors_top_10[0]

for i in range(N_top):
  if keys_deg_top[i] not in inter_list:
    color_list[keys_deg_top[i]]=colors_top_10[0]
  if keys_bet_top[i] not in inter_list:
    color_list[keys_bet_top[i]]=colors_top_10[1]

plt.figure(figsize=(15,15))
pos= nx.circular_layout(G)
nx.draw(G,pos,with_labels=True, node_color=color_list, font_size=8)

plt.legend(loc='center')
plt.show()
```

# 협업 필터링(item based collaborative filtering)

예를 들어 A 기업이 B라는 기업과 많은 거래를 했는데 B기업이 C기업과 높은 유사도를 보이면 A기업에게 C기업을 추천. 즉, A기업은 아직 C기업과의 거래를 한 적이 없기 때문에 C기업을 어떻게 평가할 것인지 예측하는 방법

- CONSIGNEE가 사용자, SHIPPER가 아이템
- 기업B와 기업C는 사용자들의 거래 분포가 비슷하므로 유사도가 높다고 판단, 기업C를 A에게 소개하게 됨
- 유사도는 코사인 유사도(cosine similarity) 사용

거래의 개수를 count로 계산하고 중복 제거

```python
da = df.groupby(['CONSIGNEE','SHIPPER']).size().reset_index()
da.columns = ['CONSIGNEE','SHIPPER', 'COUNT']

da
```

cosine similarity를 구할 때 row 기반으로 유사도를 측정하기 때문에 CONSIGNEE을 row로 만들어줌

```python
da_pi = da.pivot_table('COUNT', index = 'CONSIGNEE', columns = 'SHIPPER').fillna(0)

da_pi
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# 유사도 계산
da_sim = cosine_similarity(da_pi, da_pi)
da_sim_df = pd.DataFrame(data = da_sim, index = da_pi.index, columns = da_pi.index)

da_sim_df
```

특정 CONSIGNEE와 비교했을 때 그 CONSIGNEE와 유사한 CONSIGNEE들을 추천

```python
da_sim_df["NATIONAL PIPE COMPANY"].sort_values(ascending=False)[1:10]
```

데이터셋과 기업에 대한 거래 정보가 적어서 한계를 보임.  
예를 들어, 영화의 경우 여러 이용자가 여러 영화를 보면 볼수록 해당 알고리즘은 더 정확해짐. 그러나 한명의 이용자가 하나의 영화만 본다면 알고리즘의 정확도가 떨어짐.  
이처럼 여러 기업이 여러 기업과 거래가 이루어질수록 해당 알고리즘은 높은 정확도를 보임. 그러나 하나의 기업이 하나의 기업과 많은 거래를 하기 때문에 알고리즘의 정확도가 떨어짐

# Matrix Factorization Collaborative Filtering

행렬 분해를 이용한 협업 필터링  
대규모 다차원 행렬을 SVD와 같은 차원 감소 기법으로 분해하는 과정에서 잠재 요인(Latent Factor)를 뽑아냄. 위의 협업 필터링에서는 상품코드에 대한 정보를 추천 과정에서 고려하기 힘듬. 그러나 행렬 분해를 이용하면 행렬 데이터를 통해 잠재 요인을 뽑아낼 수 있음. 잠재 요인이 무엇이 될지는 알 수 없으나 이러한 방법으로 추천시스템이 상품코드와 같은 분야를 고려할 가능성이 커짐.

```python
from sklearn.decomposition import TruncatedSVD
from scipy.sparse.linalg import svds


db_pi = da.pivot_table('COUNT', index = 'SHIPPER', columns = 'CONSIGNEE').fillna(0)
print(db_pi.shape)

db_pi
```

R 행렬(사용자-아이템 행렬)에서 나온 값을 기반으로 latent factor score 계산 가능.   
이 값을 이용해서 해당 기업과 거래하지 않은 기업의 점수를 예측하여 추천  

SVD(Singular Value Decomposion), 특이값 분해란 m x n 크기의 데이터 행렬 A을 분해하는 알고리즘. sklearn에서 제공하는 SVD를 통해 기존 행렬 A의 성질을 거의 그대로 닮은 근사한 값을 계산할 수 있음

```python
SHIPPER_rating = db_pi.values.T
SHIPPER_rating.shape
```

15개의 component로 차원 축소

```python
SVD = TruncatedSVD(n_components=15)
matrix = SVD.fit_transform(SHIPPER_rating)
matrix.shape
```

```python
matrix[0]
```

해당 데이터를 사용해서 피어슨 상관계수 계산

```python
corr = np.corrcoef(matrix)
corr.shape
```

상관계수를 사용해서 특정 기업과 상관계수가 높은 기업 추출

```python
corr2 = corr[:200, :200]
plt.figure(figsize=(16, 10))
sns.heatmap(corr2)
```
